# Load Model and Tokenizer

```python
# s_token = "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"
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, GenerationConfig

hf_token = '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'
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Meta-Llama-3-8B-Instruct", use_auth_token=hf_token)
model = AutoModelForCausalLM.from_pretrained("meta-llama/Meta-Llama-3-8B-Instruct", device_map="auto", load_in_8bit=True, use_auth_token=hf_token)
generation_config = GenerationConfig.from_pretrained('meta-llama/Meta-Llama-3-8B-Instruct')
```

## Chat
Enter a prompt to get inference from the model

```python
while(1):
    print("\n\n")
    prompt = input("Prompt: ")
    if prompt == "exit":
      print("Exiting chat ...")
      break

    input_text = '[INST] You are a chatbot, please answer/respond to the following prompt: ' + prompt + ' ... Answer: [/INST]'

    encoded_input = tokenizer(input_text, return_tensors="pt", truncation=True, max_length=512)
    input_ids = encoded_input.input_ids.to("cuda")
    attention_mask = encoded_input.attention_mask.to("cuda")
    generation_config_dict = generation_config.to_dict()
    with torch.no_grad():
        output = model.generate(input_ids, attention_mask=attention_mask, **generation_config_dict)

    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    print("Completion:", generated_text.split('Answer: [/INST]')[-1])
```
